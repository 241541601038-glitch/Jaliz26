```python
import cv2
import matplotlib.pyplot as plt
image =cv2.imread('haikyuu.jpg')
new_width =300
new_height =200
resized_image=cv2.resize(image,(new_width,new_height))
plt.figure(figsize=(50,50))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(cv2.cv+color(image,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(1,2,3)
plt.title('Resized image')
plt.imshow(cv2.cv+color(resized_image,cv2.COLOR_BGR2RGB))
plt.axis('off')+
plt.show()
```
