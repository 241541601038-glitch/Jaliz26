```python
selling_price = float(input("Enter selling price: "))
gst_rate = float(input("Enter GST rate (%): "))

cgst = (selling_price * gst_rate) / 200
sgst = (selling_price * gst_rate) / 200
total_tax = cgst + sgst
total_amount = selling_price + total_tax

print("\n-------------Invoice-------------")
print(f"MALL Name:PHOENIX MALL")
print(f"Item Name: BOAT Airdopes")
print(f"Price: ₹{selling_price}")
print(f"GST Rate: {gst_rate}%")
print(f"CGST ({gst_rate/2}%): ₹{cgst:.2f}")
print(f"SGST ({gst_rate/2}%): ₹{sgst:.2f}")
print(f"Total Amount: ₹{total_amount:.2f}")   
```
